```python
import math

# this is the current state of the game 
board = [
    ['O', 'X', 'O'],
    ['X', 'O', ' '],
    ['X', ' ', ' ']
]

# Function to print the board
def print_board(b):
    for row in b:
        print(" | ",row[0], " | ",row[1], " | ", row[2], " | ")
        print("--------------------")
    print("   ------*------    ")

# this function is to check for the winner
def check_winner(b):

    win_conditions = [
        [b[0][0], b[0][1], b[0][2]],
        [b[1][0], b[1][1], b[1][2]],
        [b[2][0], b[2][1], b[2][2]],
        [b[0][0], b[1][0], b[2][0]],
        [b[0][1], b[1][1], b[2][1]],
        [b[0][2], b[1][2], b[2][2]],
        [b[0][0], b[1][1], b[2][2]],
        [b[0][2], b[1][1], b[2][0]],
    ]
    if ['X', 'X', 'X'] in win_conditions:
        return 'X'
    if ['O', 'O', 'O'] in win_conditions:
        return 'O'
    return None

# Check for a draw
def is_draw(b):
    for row in b:
        if ' ' in row:
            return False
    return True

# alphaBeta function with alpha-beta pruning
def alphaBeta(b, depth, is_maximizing, alpha, beta):
    winner = check_winner(b)
    if winner == 'X':
        return 10 - depth
    if winner == 'O':
        return depth - 10
    if is_draw(b):
        return 0

    if is_maximizing:
        max_eval = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    eval = alphaBeta(b, depth + 1, False, alpha, beta)
                    b[i][j] = ' '
                    max_eval = max(max_eval, eval)
                    alpha = max(alpha, eval)
                    if beta <= alpha:
                        break
        return max_eval
    else:
        min_eval = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    eval = alphaBeta(b, depth + 1, True, alpha, beta)
                    b[i][j] = ' '
                    min_eval = min(min_eval, eval)
                    beta = min(beta, eval)
                    if beta <= alpha:
                        break
        return min_eval

# Function to find the best move for 'X'
def find_best_move(b):
    best_val = -math.inf
    best_move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if b[i][j] == ' ':
                b[i][j] = 'X'
                move_val = alphaBeta(b, 0, False, -math.inf, math.inf)
                b[i][j] = ' '
                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val
    return best_move

# Main game loop
while True:
    print_board(board)
    if check_winner(board) == 'X':
        print("X wins!")
        break
    elif check_winner(board) == 'O':
        print("O wins!")
        break
    elif is_draw(board):
        print("It's a draw!")
        break

    # Find and make the best move for 'X'
    x, y = find_best_move(board)
    if x != -1 and y != -1:
        board[x][y] = 'X'
    else:
        break
```

```python


def showBoard(board):
    print('--------------')
    for i in range(0, 9, 3):
        print('|', board[i], '|', board[i + 1], '|', board[i + 2], '|')
        print('--------------')

def checkWinner(board):
    """Checks if there's a winner or if the game is a draw."""
    win_conditions = [(0, 1, 2), (3, 4, 5), (6, 7, 8),(0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6)]           
    for (a, b, c) in win_conditions:
        if board[a] == board[b] == board[c] and board[a] != ' ':
            return board[a]  # Return the winner ('X' or 'O')

    if ' ' not in board:
        return "Draw"  # If the board is full, it's a draw

    return None  # No winner yet

def alphaBeta(player, board, alpha=-float('inf'), beta=float('inf')):
    """alpha-beta pruning to decide the optimal move."""
    result = checkWinner(board)
    
    # Terminal states
    if result == 'X':
        return None, 1  # Maximizing player wins
    elif result == 'O':
        return None, -1  # Minimizing player wins
    elif result == "Draw":
        return None, 0  # Draw

    if player == 'X':
        max_value = -float('inf')
        best_move = None
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'X'
                _, score = alphaBeta('O', board, alpha, beta)
                board[i] = ' '
                if score > max_value:
                    max_value = score
                    best_move = i
                alpha = max(alpha, max_value)
                if beta <= alpha:
                    break
        return best_move, max_value
    else:
        min_value = float('inf')
        best_move = None
        for i in range(9):
            if board[i] == ' ':
                board[i] = 'O'
                _, score = alphaBeta('X', board, alpha, beta)
                board[i] = ' '
                if score < min_value:
                    min_value = score
                    best_move = i
                beta = min(beta, min_value)
                if beta <= alpha:
                    break
        return best_move, min_value

def playtictactoe(board):
    turn = 'X'
    while True:
        if turn == 'X':
            while True:
                try:
                    move = int(input("Choose a position (1-9): ")) - 1
                    if board[move] == ' ':
                        board[move] = 'X'
                        break
                    else:
                        print("That spot is already taken, try again.")
                except (ValueError, IndexError):
                    print("Invalid input, choose a number from 1 to 9.")
        else:
            move, _ = alphaBeta('O', board)
            if move is not None:
                board[move] = 'O'

        # Display board after each move
        showBoard(board)

        # Check game status
        status = checkWinner(board)
        if status == 'X':
            print("You win!")
            break
        elif status == 'O':
            print("AI wins!")
            break
        elif status == "Draw":
            print("It's a draw!")
            break

        # Switch turn
        turn = 'O' if turn == 'X' else 'X'

# Initialize an empty board and start the game
empty_board = [' ' for _ in range(9)]
showBoard(empty_board)
playtictactoe(empty_board)
```
