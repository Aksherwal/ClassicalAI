## Reinforcement learning in Tic-Tac-Toe

```python
import random

# Dictionary to store states and their values
saved_states = {}

# Learning rate
learning_rate = 0.2

def printBoard(X, O):
    board = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ']
    for i in X:
        board[i] = 'X'
    for j in O:
        board[j] = 'O'
    print('--------------')
    print('|', board[0], '|', board[1], '|', board[2], '|')
    print('--------------')
    print('|', board[3], '|', board[4], '|', board[5], '|')
    print('--------------')
    print('|', board[6], '|', board[7], '|', board[8], '|')
    print('--------------')

def checkWinner(lst):
    winning_cases = [(0, 1, 2), (3, 4, 5), (6, 7, 8), (0, 3, 6), (1, 4, 7), (2, 5, 8), (0, 4, 8), (2, 4, 6)]
    for case in winning_cases:
        if case[0] in lst and case[1] in lst and case[2] in lst:
            return True
    return False

def giveRandomPosition(lst):
    x = random.randint(0, 8)
    if x in lst:
        return giveRandomPosition(lst)
    return x

# Function to get state tuple
def get_state(X, O):
    return (tuple(sorted(X)), tuple(sorted(O)))

# Function to update state values with learning rate
def update_state_values(states, outcome):
    global saved_states
    for state in reversed(states):
        if state not in saved_states:
            saved_states[state] = 0  # Initialize state value if not seen before
        # Apply the learning rate to adjust state value
        saved_states[state] += learning_rate * (outcome - saved_states[state])
        outcome = saved_states[state]  # Set current state's value for the next state

def tic_tac_toe():
    X = []
    O = []
    elements_filled = 0
    player = 'X'  # First player is X
    states_visited = []  # List of visited states
    while elements_filled < 9:
        x = giveRandomPosition(X + O)
        if player == 'X':
            X.append(x)
            states_visited.append(get_state(X, O))  # Record the state
            player = 'O'
            elements_filled += 1
            printBoard(X, O)
            if checkWinner(X):
                print('X Wins!')
                update_state_values(states_visited, 1)  # Positive reward for X win
                return 1  # X wins
        else:
            O.append(x)
            states_visited.append(get_state(X, O))  # Record the state
            player = 'X'
            elements_filled += 1
            printBoard(X, O)
            if checkWinner(O):
                print('O Wins!')
                update_state_values(states_visited, -1)  # Negative reward for O win
                return -1  # O wins
    print("It's a Draw!")
    update_state_values(states_visited, 0)  # Neutral reward for draw
    return 0  # Draw

# Play multiple games to learn
for _ in range(100):  # Run the game 100 times for learning
    tic_tac_toe()

# Print learned states and their values
print("Learned State Values:")
for state, value in saved_states.items():
    print(state, ":", value)
```
